```python
with open("outputs/transcribed_directly_intermediate.txt", "r") as file:
    content = file.read()

# have some replacements to do, structured as a dictionary
replacements = {
    "old_string_1": "new_string_1",
    "old_string_2": "new_string_2",
    "old_string_3": "new_string_3",
}
for old, new in replacements.items():
    content = content.replace(old, new)

with open("outputs/transcribed_directly_intermediate.txt", "w") as file:
    file.write(content)
```
